```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Configuramos el estilo de los gráficos
sns.set_theme(style="whitegrid")

# 1. Carga de datos 
df_listings = pd.read_csv('../Data/listings.csv')

# Un primer vistazo a las primeras filas
display(df_listings.head())

# 2. Exploración sistemática sugerida 
print("--- 1. Información General del Dataset ---")
display(df_listings.info())

print("\n--- 2. Estadísticas Descriptivas ---")
display(df_listings.describe())

print("\n--- 3. Conteo de Valores Nulos ---")
display(df_listings.isnull().sum())
```

```python
# 1. Limpieza express de la columna 'price' para poder graficar
if df_listings['price'].dtype == 'O':
    df_listings['price'] = df_listings['price'].str.replace('$', '', regex=False)
    df_listings['price'] = df_listings['price'].str.replace(',', '', regex=False)
    df_listings['price'] = df_listings['price'].astype(float)

# 2. Histograma: Distribución general de los precios
# Filtramos los precios mayores a 1000 temporalmente solo para que el gráfico no se deforme por los outliers extremos
plt.figure(figsize=(10, 6))
sns.histplot(df_listings[df_listings['price'] < 1000]['price'], bins=50, kde=True, color='teal')
plt.title('Distribución de Precios por Noche (menores a $1000)')
plt.xlabel('Precio ($)')
plt.ylabel('Cantidad de Propiedades')
plt.show()

# 3. Boxplot: Precios según el tipo de habitación (Room Type)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_listings[df_listings['price'] < 1000], x='room_type', y='price', palette='Set2')
plt.title('Dispersión de Precios por Tipo de Habitación')
plt.xlabel('Tipo de Habitación')
plt.ylabel('Precio ($)')
plt.show()
```

```python
# 4. Heatmap de Correlación: Relación entre precio, características y calificaciones
# Seleccionamos un subconjunto de columnas numéricas clave
columnas_clave = ['price', 'accommodates', 'bedrooms', 'beds', 'number_of_reviews', 'review_scores_rating']

# Calculamos la matriz de correlación (ignora los nulos automáticamente)
matriz_correlacion = df_listings[columnas_clave].corr()

# Graficamos el mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor: Correlación de Variables Numéricas')
plt.show()
```

```python
# 4. Heatmap de Correlación (SIN OUTLIERS y con más variables)

# Filtramos las propiedades con precios extremos para que no rompan la matemática
df_filtrado = df_listings[df_listings['price'] < 1000]

# Agregamos bathrooms, guests_included y availability_365
columnas_clave = [
    'price', 
    'accommodates', 
    'bedrooms', 
    'bathrooms', 
    'beds', 
    'guests_included',
    'availability_365',
    'number_of_reviews', 
    'review_scores_rating'
]

# Calculamos la matriz de correlación SOBRE EL DATASET FILTRADO
matriz_correlacion = df_filtrado[columnas_clave].corr()

# Graficamos
plt.figure(figsize=(10, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)
plt.title('Mapa de Calor: Correlación (Precios < $1000)')
plt.show()
```

Análisis Exploratorio de Datos (EDA) - Hallazgos Iniciales
Caso de Negocio: Airbnb

1. Carga de Datos y Limpieza Preliminar
El análisis inicial se realizó sobre el dataset principal listings.csv.

Estructura y Nulos: A través de la función df.info() y df.isnull().sum(), se detectó una gran cantidad de valores nulos, principalmente en columnas de texto libre o descriptivas como space (7.194 nulos) y neighborhood_overview (7.665 nulos). Esto indica la necesidad de descartar o imputar estas variables cualitativas en la siguiente etapa, priorizando las variables cuantitativas y categóricas duras.

Transformación Crítica: La variable objetivo del análisis, price, se encontraba almacenada como texto (string) con símbolos monetarios y separadores de miles (ej: $1,200.00). Fue necesario realizar una limpieza express, eliminando estos caracteres y convirtiendo la columna a tipo numérico (float) para poder realizar operaciones estadísticas y gráficas.

2. Distribución General de Precios (Análisis Univariado)
Para entender cómo se distribuye la oferta, se generó un Histograma filtrando temporalmente los valores extremos (precios mayores a $1000) para no distorsionar la visualización.

Hallazgo: La distribución presenta un sesgo hacia la derecha (right-skewed). Se observa un fuerte comportamiento psicológico por parte de los anfitriones a establecer tarifas en números redondos, generando picos muy marcados en valores como $600, $700, $800 y $1000.

Concentración: La gran mayoría de la oferta se concentra en los rangos medios y altos dentro del umbral de los $1000, con muy pocas opciones disponibles por debajo de los $200.

3. Dispersión de Precios por Tipo de Habitación (Análisis Bivariado)
Se utilizó un Boxplot para analizar la relación entre el precio y la variable categórica room_type.

Categoría más costosa: Entire home/apt (Casa/departamento entero) no solo tiene la mediana de precio más alta (cercana a los $800), sino que también presenta la caja más alargada. Esto demuestra una altísima variabilidad de precios en esta categoría, que abarca desde propiedades estándar hasta opciones de lujo.

Categorías accesibles: Las opciones de Shared room (Habitación compartida) y Private room (Habitación privada) son claramente las más económicas y presentan una dispersión mucho menor.

Presencia de Outliers: Se detectaron múltiples valores atípicos (puntos fuera de los "bigotes" del gráfico) en casi todas las categorías, lo que requerirá un tratamiento estadístico formal en la etapa de limpieza para no sesgar los promedios generales.

4. Correlación de Variables Numéricas (Análisis Multivariado)
Para responder a la pregunta de negocio sobre qué características hacen rentables a las propiedades, se construyó un Mapa de Calor (Heatmap) cruzando el precio con variables físicas (accommodates, bedrooms, beds, bathrooms) y de satisfacción (review_scores_rating, number_of_reviews).

Hallazgo Clave (La paradoja del tamaño): Sorprendentemente, la correlación lineal entre el precio y el tamaño físico de la propiedad (ej. accommodates o bedrooms) es extremadamente baja (cercana a 0.07 y 0.01 respectivamente). Lo mismo ocurre con las métricas de satisfacción.

Justificación: Esto demuestra que el precio en Airbnb no sigue una lógica puramente aditiva o lineal. Al analizar el dataset completo, se mezclan tipos de propiedades y ubicaciones de valor dispar. Un monoambiente (Entire home/apt) en un barrio de alta demanda turística puede tener un precio muy superior al de una casa compartida en la periferia, sin importar que esta última tenga más camas o mejor calificación.

Conclusión: El precio es una variable multivariada fuertemente determinada por variables categóricas (tipo de habitación y ubicación geográfica), las cuales "rompen" la correlación lineal de las características físicas al ser analizadas en conjunto.
