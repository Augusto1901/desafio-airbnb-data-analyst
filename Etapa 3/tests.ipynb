```python
# Visualización de la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(df_test['price'], bins=50, kde=True)
plt.title('Distribución de Precios de Airbnb')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
  # Limitar el eje x para una mejor visualización
plt.show()
```

```python
# grafico interactivo de los precios
import plotly.express as px
fig = px.histogram(df_test, x='price', nbins=100, title='Distribución de Precios de Airbnb')
fig.update_layout(xaxis_title='Precio', yaxis_title='Frecuencia')
fig.show()
```
